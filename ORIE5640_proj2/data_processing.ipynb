# Data preparation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('DTB3.csv', index_col='DATE', parse_dates=True)
data['DTB3'] = data['DTB3'].replace('.', np.nan).astype(float).fillna(method='ffill')
log_ret = np.log(data['DTB3'] / data['DTB3'].shift(1)).dropna()

plt.plot(log_ret)
plt.show()
```
