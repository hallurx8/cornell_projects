```python
import yfinance as yf
import datetime as dt
import numpy as np
import pandas as pd

# Define the time range
end_date = dt.date(2017,1,1)
start_date = dt.date(2011,6,1)
```

```python
GD = yf.download('GD=F', start=start_date, end=end_date, interval='1wk')
r_INDEX = pd.DataFrame()
r_INDEX['GD'] = GD['Close'].pct_change()
r_INDEX.to_pickle('index_ret.pkl')
  
```

```python
future_price = pd.DataFrame()
future_price["GC"] = GC['Close']
future_price["ZC"] = ZC['Close']
future_price["HE"] = HE['Close']
future_price["CT"] = CT['Close']
future_price["YM"] = YM['Close']
# future_price["ZT"] = ZT['Close']
# future_price["ZF"] = ZF['Close']
future_price["NG"] = NG['Close']
future_price.head(10)
```

```python
df_P_bear = future_price
mask = pd.date_range(start=df_P_bear.index[4], end=df_P_bear.index[-1], freq='5D')
df = df_P_bear[df_P_bear.index.isin(mask)]
df.head()
```

```python
df.to_pickle('future_price_bear.pkl')
```
